# object oriented plotting approach

1.   Create a bucket that will contain the plot(figure)
2.  Fill the bucket with the plot we want-axes(ax)


```python
# Trying to show the data
x = df['Time (hours)']
y1 = df['Runner 1 Distance']
y2 = df['Runner 2 Distance']
```

```python
fig, ax = plt.subplots()
ax.plot(x,y1);
# the semi-colon removes the object address
```

```python
# lets add another line to the figure
fig, ax = plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2);
```

```python
# same thing applies to the scatter plot
fig, ax = plt.subplots()
ax.scatter(x, y1)
ax.scatter(x, y2)
```

```python
# plotting the average of each as a bar
runner_1_mean = y1.mean()
runner_2_mean = y2.mean()
bar_labels= ['Runner 1', 'Runner 2']
fig, ax = plt.subplots()
ax.bar(bar_labels, [runner_1_mean, runner_2_mean]);
```

### Adding labels and legend
We can use the 'label' argument in ax.plot, ax.bar, ax.scatter to provide a text label for the data. to see the labels, we need to add 'legend with ax.legend()

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label= 'Runner 1')
ax.plot(x,y2, label= 'Runner 2')
ax.legend();

# adding axis labels and titles
ax.set_ylabel("Distance (miles)")
ax.set_xlabel("Time (hours)")
ax.set_title("Comparing runner distances over Time")
```

```python
# say we are lazy(like most devs) and want to write the whole setting
# wagwes in one line of code
fig, ax = plt.subplots()
ax.scatter(x,y1, label = "runner 1")
ax.scatter(x,y2, label = "runner 2")
ax.legend()
ax.set(ylabel= "Distance (miles)", xlabel= "Time (hours)", title= "Comparing runner Distances over Time");
```

### Histograms
categorical features strive here
Histograms work by creating bins that consist of bins. it then counts the frequency of occurence of each bin.

```python
import pandas as pd
import seaborn as sns
import numpy as np
df.columns
```

```python
# considering the salePrice column
df['SalePrice'].hist()
# this returns an axes object that we can modify as seen previously
```

```python
# object based option
ax = df["SalePrice"].hist()

# remove scientific notation.
ax.ticklabel_format(style= 'plain')

# rotate your xticklabels
ax.tick_params(axis= 'x', rotation = 45)
```

```python
#  trying to find all the unique values in a column
df['PID'].nunique()
# df['PID'].info()
```

```python
# changing the number of bins to 30. more bins, more detail
# ax = df['SalePrice'].hist(bins= 50)
ax = df['SalePrice'].hist(bins= 'auto') # the library chooses the bins for me
ax.ticklabel_format(style= 'plain')
ax.tick_params(axis= 'x', rotation= 45)
```

```python
# add title, labels, edges around each bin for better visibility
# changing the number of bins changes the frequency distribution
ax= df['SalePrice'].hist(bins= 45, edgecolor= 'black')
ax.tick_params(axis= 'x', rotation= 45)
ax.ticklabel_format(style= 'plain')
ax.set(ylabel= 'Number of houses', xlabel= 'Price ($)', title= "Distribution of house Prices");
```

# Using seaborn's sns.histplot
it is used to create histograms

```python
# use seaborn's histplot function and save the ax
ax = sns.histplot(data = df, x = 'SalePrice')
ax.ticklabel_format(style= 'plain')
ax.tick_params(axis= 'x', rotation = 45)
```

## advantages of seaborn
Seaborn's plotting functions generally have many convinient parameters that make it easy to modify the resulting visualization without additional work

eg. manual creation of the size of each  bin is a real issue in matplotlib/pandas
We dont suffer like this in seaborn(we specify using the bin_width parameter)

```python
# using binwidth arg to make histogram with $50k bins
ax = sns.histplot(data= df, x = 'SalePrice', binwidth= 50_000)
ax.ticklabel_format(style= "plain")
ax.tick_params(axis= 'x', rotation = 45)
```

```python
# how can we change the scale of our data?
ax = sns.histplot(data = df, x= 'SalePrice')
ax.ticklabel_format(style= "plain")
ax.tick_params(axis= 'x', rotation= 45)

# changing the scale of our visualization and not that of our original data
ax.set_yticks(range(0,400,100))
```

```python
# getting a portion of the entire dataset by drilling down to it
price_filter = (df['SalePrice']>= 500000) & (df['SalePrice']<= 750000)
ax = sns.histplot(data= df[price_filter], x = 'SalePrice',binwidth= 50_000)
ax.ticklabel_format(style= 'plain')
ax.tick_params(axis = 'x', rotation = 45)
```

# Advantages of seaborn's histplot


1.   has many convinient parameters that enable easier modification of visualizations.




```python
# using bin_width arg to make a histogram with 50k bins and plot
# as % of total
# we can also use probability, density and so on
ax = sns.histplot(data= df, x= 'SalePrice', binwidth= 50_000, stat= 'percent')
ax.ticklabel_format(style= 'plain')
ax.tick_params(axis= 'x',rotation= 45);
```
